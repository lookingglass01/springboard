With the data wrangling done, it's time to see what story the data tells. In order to do this, all of the categorical variables will be plotted.

Import libraries and dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv (r"C:\Users\Laura-Black\Documents\PhD\Data Scientist\Springboard\Capstone Projects\Drug Abuse ED Visits\Output files\Capstone1DataWrangling.csv", 
                     index_col=0, dtype={"Buyer_Part_Number": str}, low_memory=False)

print(df.head())
print(df.head())

```

```python
print(df.info())
print(list(df.columns.values))
```

Plot Metro distribution

```python
plt.figure(figsize=(16,10))
sns.countplot(x='METRO',data=df)
plt.xticks(rotation=45)
plt.title('Metro Area', fontsize=30)
plt.xlabel('Metro Area', fontsize=20)
plt.ylabel('Number of Cases', fontsize=20)
```

Plot Gender distribution

```python
plt.figure(figsize=(12,8))
sns.countplot(x='SEX',data=df)
plt.xticks(rotation=45)
plt.title('Gender', fontsize=30)
plt.xlabel('Gender', fontsize=20)
plt.ylabel('Number of Cases', fontsize=20)
```

Plot Race distribution

```python
plt.figure(figsize=(12,8))
sns.countplot(x='RACE',data=df)
plt.xticks(rotation=45)
plt.title('Race', fontsize=30)
plt.xlabel('Race', fontsize=20)
plt.ylabel('Number of Cases', fontsize=20)
```

Plot Age Category distribution

```python
plt.figure(figsize=(16,10))
counts = df["AGECAT"].value_counts().sort_index()
colors = plt.cm.tab10(np.arange(len(counts)))
ax = counts.plot.bar(color=colors)
plt.xticks(rotation=45)
plt.title('Age Category', fontsize=30)
plt.xlabel('Age Category', fontsize=20)
plt.ylabel('Number of Cases', fontsize=20)
```

Plot Quarter distribution

```python
plt.figure(figsize=(12,8))
sns.countplot(x='QUARTER',data=df)
plt.title('Quarter of Year', fontsize=30)
plt.xlabel('Quarter of Year', fontsize=20)
plt.ylabel('Number of Cases', fontsize=20)
```

Plot Time of Day distribution

```python
plt.figure(figsize=(12,6))
order = ["12:00-5:59 AM","6:00-11:59 AM","12:00-5:59 PM","6:00-11:59 PM"]
counts = df["DAYPART"].value_counts().sort_index().loc[order]
colors = plt.cm.tab10(np.arange(len(counts)))
ax = counts.plot.bar(color=colors)
plt.xticks(rotation=45)
plt.title('Time of Day', fontsize=30)
plt.xlabel('Time of Day', fontsize=20)
plt.ylabel('Number of Cases', fontsize=20)
```
