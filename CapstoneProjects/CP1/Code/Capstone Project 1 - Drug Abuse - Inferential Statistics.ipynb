Now that we've seen the story that the data tells, it's time to perform some hypothesis testing.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats as st
import statsmodels as sm

df = pd.read_csv (r"C:\Users\Laura-Black\Documents\PhD\Data Scientist\Springboard\Capstone Projects\Drug Abuse ED Visits\Output files\Capstone1DataWrangling.csv", 
                     index_col=0, low_memory=False)

print(df.head())
print(df.head())
```

First, the locations of the cases will be analyzed. Is there a significant difference in the frequency of cases in each of the cities?

Ho: p1 = p2... = p15
Ha: p1 != p2... != p15

```python
nobs =  len(df['METRO'])
table1 = df['METRO'].value_counts()
print(table1)

model1 = sm.stats.proportion.proportions_chisquare(table1, nobs=nobs)
print(model1)
stat1 = 96116.77852955775
p1 = 0.0

print('\n stat = ', stat1)
print('\n p-value = ', p1)

# interpret test-statistic
prob = 0.95
dof1 = nobs - 1
critical1 = st.chi2.ppf(prob, dof1)
print('\n probability=%.3f, critical=%.3f, dof=%.3f' % (prob, critical1, dof1))

# interpret p-value
alpha = 1.0 - prob
print('\n significance=%.3f, p=%.3f' % (alpha, p1))
if p1 <= alpha:
    print('\n Reject H0')
else:
    print('\n Fail to reject H0')
```

Next, the race of the cases will be analyzed. Does the race of the patient vary signifcantly?

Ha: p1 != p2... != p15

```python
nobs =  len(df['RACE'])
table1 = df['RACE'].value_counts()
print(table1)

model2 = sm.stats.proportion.proportions_chisquare(table1, nobs=nobs)
print(model2)
stat2 = 167467.57531270315
p2 = 0.0

print('\n stat = ', stat2)
print('\n p-value = ', p2)

# interpret test-statistic
prob = 0.95
dof2 = nobs - 1
critical2 = st.chi2.ppf(prob, dof2)
print('\n probability=%.3f, critical=%.3f, dof=%.3f' % (prob, critical2, dof2))

# interpret p-value
alpha = 1.0 - prob
print('\n significance=%.3f, p=%.3f' % (alpha, p2))
if p1 <= alpha:
    print('\n Reject H0')
else:
    print('\n Fail to reject H0')
```
